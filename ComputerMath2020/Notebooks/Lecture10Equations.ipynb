åˆ™ï¼Œ1å¹´åï¼Œåœ¨åŸå¸‚å’ŒéƒŠåŒºç”Ÿæ´»çš„äººå£æ¯”ä¾‹å¯ç”±$x_1 = A x_0$ æ±‚å¾—ã€‚2å¹´åçš„æ¯”ä¾‹å¯ç”±$x_2 = Ax_1 = A^2 x_0$æ±‚å¾—ã€‚ä¸€èˆ¬åœ°ï¼Œnå¹´åçš„æ¯”ä¾‹å¯ç”±$x_n = A^n x_0$ç»™å‡ºã€‚å¦‚æœè®¡ç®—n=10ï¼Œ30å’Œ50æ—¶çš„ç™¾åˆ†æ¯”ï¼Œå¹¶å°†å®ƒä»¬èˆå…¥åˆ°æœ€æ¥è¿‘çš„ç™¾åˆ†æ¯”ï¼Œæˆ‘ä»¬æœ‰ï¼š

$x_{10}=A^{10} x_0 = \begin{bmatrix} 0.272 \\ 0.728 \end{bmatrix},

ä¸Šå¼ä¸­çŸ©é˜µä¹˜æ³•çš„éƒ¨åˆ†ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨Pythonæ¥ç¼–ç¨‹å®ç°ï¼š

```python
import numpy as np
# np.set_printoptions(formatter={'float': '{: 0.2f}'.format})

A0 = np.array([[0.94,0.02],[0.06,0.98]])
x = np.array([[0.3],[0.7]])

n = 50
years = [1,2,10,30,50]

A = A0
for year in range(n):  # yearçš„å–å€¼èŒƒå›´æ˜¯[0:49],å…¶ä¸­0ä¸ºç¬¬ä¸€å¹´ï¼Œ49ä¸ºç¬¬50å¹´
    res = np.dot(A, x)    
    A = np.dot(A, A0)
    if year+1 in years:
        print('{:2d}å¹´åçš„å±…ä½æ¯”ä¾‹ä¸ºï¼š{}'.format(year+1, res.T))
```

> 2. ç»™å®šä¸€ä¸ªå‘é‡$c_1 u_1 + c_2 u_2$ï¼Œæ±‚å®ƒåœ¨$e_1, e_2$ä¸‹çš„åæ ‡ã€‚

**ã€ä¹ é¢˜5.13ã€‘** ä»¤$u_1=(1,1,1)^T, u_2=(1,2,2)^T, u_3=(2,3,4)^T$ï¼Œæ±‚ï¼š

2) æ±‚ä¸‹åˆ—å‘é‡åœ¨åŸº$[u_1, u_2, u_3]$ä¸‹çš„åæ ‡ã€‚(i) $A=(3,2,5)^T$, (ii) $B=(1,1,2)^T$, (iii) $C=(2,3,2)^T$ 

> é—®é¢˜åˆ†æï¼šä»æ ‡å‡†åŸº$[e_1, e_2, e_3]$åˆ°ç‰¹å®šåŸº$[u_1, u_2, u_3]$çš„åŸºåº•å˜æ¢ï¼ˆåæ ‡å˜æ¢ï¼‰ã€‚æ ¹æ®åæ ‡å˜æ¢å…¬å¼$c=U^{-1} x$å³å¯æ±‚å¾—åæ ‡å’Œè½¬ç§»çŸ©é˜µã€‚

$U^{-1}=\begin{bmatrix} 1 & 1 & 2\\ 1 & 2 & 3 \\ 1 & 2 & 4 \end{bmatrix} ^{-1}$

$[A]_u = U^{-1} x = \begin{bmatrix} 1 & 1 & 2\\ 1 & 2 & 3 \\ 1 & 2 & 4 \end{bmatrix} ^{-1} \begin{bmatrix} 3 \\ 2 \\ 5 \end{bmatrix}$

$[B]_u = U^{-1} x = \begin{bmatrix} 1 & 1 & 2\\ 1 & 2 & 3 \\ 1 & 2 & 4 \end{bmatrix} ^{-1} \begin{bmatrix} 1 \\ 1 \\ 2 \end{bmatrix}$

$[C]_u = U^{-1} x = \begin{bmatrix} 1 & 1 & 2\\ 1 & 2 & 3 \\ 1 & 2 & 4 \end{bmatrix} ^{-1} \begin{bmatrix} 2 \\ 3 \\ 2 \end{bmatrix}$

åç»­å·¥ä½œäº¤ç»™Pythonæ¥å®ç°ã€‚


```python
import numpy as np
from scipy import linalg

U = np.array([[1,1,2],[1,2,3],[1,2,4]])
A = np.array([[3],[2],[5]])
B = np.array([[1],[1],[2]])
C = np.array([[2],[3],[2]])

U_inv = linalg.inv(U)

print('U_inv=\n {}'.format(U_inv))
print('Au=\n{}'.format(np.dot(U_inv, A)))
print('Bu=\n{}'.format(np.dot(U_inv, B)))
print('Cu=\n{}'.format(np.dot(U_inv, C)))
```

**ã€ä¹ é¢˜5.16ã€‘** ç»™å®š$v_1=\begin{bmatrix} 2 \\ 6 \end{bmatrix}, v_2=\begin{bmatrix} 1 \\ 4 \end{bmatrix}, S = \begin{bmatrix} 4 & 1 \\ 2 & 1 \end{bmatrix}$ï¼Œæ±‚ï¼š$u_1, u_2$ï¼Œä½¿å¾—$S$ä¸ºä»$[v_1, v_2]$åˆ°$[u_1, u_2]$çš„è½¬ç§»çŸ©é˜µã€‚
