### **ã€ä¾‹6.7ã€‘ ç½‘ç»œçš„çŸ©é˜µåˆ†å‰²å’Œè¿æ¥**

ç»™å‡ºå¦‚ä¸‹ç”µè·¯å›¾ï¼ŒåŒæ—¶å·²çŸ¥è¾“å…¥ç”µæµ$i_{in}$å’Œè¾“å…¥ç”µå‹$u_{in}$ï¼Œå¹¶ä¸”å·²çŸ¥å››ä¸ªç”µé˜»çš„ç”µé˜»å€¼ï¼Œæ±‚è¾“å‡ºç”µæµ$i_{out}$å’Œè¾“å‡ºç”µå‹$u_{out}$ã€‚

ä¸ºäº†æ±‚è¾“å‡ºç”µæµã€ç”µå‹ï¼Œæˆ‘ä»¬å¯ä»¥å°†ä»¥ä¸Šç”µè·¯å›¾æŠ½è±¡åŒ–ä¸ºé»‘ç›’å›¾ï¼Œå¹¶æ‹†åˆ†æˆä¸åŒç”µè·¯æ¨¡å—çš„ç»„åˆï¼Œå†é€šè¿‡çŸ©é˜µä¹˜æ³•æ±‚è§£æœ€ç»ˆçš„è¾“å‡ºçŸ©é˜µã€‚å¦‚ä¸‹å›¾æ‰€ç¤ºï¼Œå…¶ä¸­$A1ã€A4$ä¸ºä¸²è”ç”µè·¯ï¼Œ$A2ã€A3$ä¸ºå¹¶è”ç”µè·¯ã€‚ï¼š

**1. æ±‚ç¬¬ä¸€ä¸ªå­ç½‘ç»œï¼ˆä¸²è”ç”µè·¯ï¼‰çš„ä¼ è¾“çŸ©é˜µ**

1). æ ¹æ®ç”µè·¯åŸç†å¯ä»¥å¾—åˆ°

$i_2 = i_1, u_2 = u_1 - i_1 R_1 $

2). å°†å…¬å¼è½¬æ¢ä¸ºçŸ©é˜µæ–¹ç¨‹

$\begin{bmatrix} u_2 \\ i_2 \end{bmatrix}
=\begin{bmatrix} u_1 - i_1 R_1 \\ 0 - i_1 \end{bmatrix}
=\begin{bmatrix} 1 & -R_1 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}
=A_1 \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}$

3). è·å¾—ç¬¬ä¸€ä¸ªå­ç½‘ç»œçš„ä¼ è¾“çŸ©é˜µï¼Œå³`å•ä¸€ä¸²è”ç”µè·¯`çš„ä¼ è¾“çŸ©é˜µã€‚

$A_1=\begin{bmatrix} 1 & -R_1 \\ 0 & 1 \end{bmatrix}$

**2. æ±‚ç¬¬äºŒä¸ªå­ç½‘ç»œï¼ˆå¹¶è”ç”µè·¯ï¼‰çš„ä¼ è¾“çŸ©é˜µ**

1). æ ¹æ®ç”µè·¯åŸç†å¯ä»¥å¾—åˆ°

$i_3 = i_2 - u_2/R_2, u_3 = u_2$

2). å°†å…¬å¼è½¬æ¢ä¸ºçŸ©é˜µæ–¹ç¨‹

$\begin{bmatrix} u_3 \\ i_3 \end{bmatrix}
=\begin{bmatrix} u_2 + 0 i_2 \\ -u_2/R_2 + i_2 \end{bmatrix}
=\begin{bmatrix} 1 & 0 \\ -1/R_2 & 1 \end{bmatrix} \begin{bmatrix} u_2 \\ i_2 \end{bmatrix}
=A_2 \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}$

3). è·å¾—ç¬¬äºŒä¸ªå­ç½‘ç»œçš„ä¼ è¾“çŸ©é˜µï¼Œå³`å•ä¸€å¹¶è”ç”µè·¯`çš„ä¼ è¾“çŸ©é˜µã€‚

$A_2=\begin{bmatrix} 1 & 0 \\ -1/R_2 & 1 \end{bmatrix}$

**3. æ±‚æ€»ç”µè·¯çš„ä¼ è¾“çŸ©é˜µ**

1). è”ç«‹ä»¥ä¸Šä¸¤ä¸ªç½‘ç»œè·å¾—è¾“å…¥è¾“å‡ºçš„å…³ç³»

$\begin{bmatrix} u_3 \\ i_3 \end{bmatrix}
=A_2 \begin{bmatrix} u_2 \\ i_2 \end{bmatrix}
=A_2 A_1 \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}$

2). è·å¾—æ€»ç”µè·¯ä¼ è¾“çŸ©é˜µ$ğ‘¨$

$A=A_2 A_1
=\begin{bmatrix} 1 & 0 \\ -1/R_2 & 1 \end{bmatrix} \begin{bmatrix} 1 & -R_1 \\ 0 & 1 \end{bmatrix}
=\begin{bmatrix} 1 & -R_1 \\ -1/R_2 & 1+R_1/R_2 \end{bmatrix}$

3). åˆ©ç”¨ä¼ è¾“çŸ©é˜µæ±‚è¾“å‡ºç”µå‹å’Œè¾“å‡ºç”µæµ

$\begin{bmatrix} u_3 \\ i_3 \end{bmatrix}
=A \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}
=\begin{bmatrix} 1 & -R_1 \\ -1/R_2 & 1+R_1/R_2 \end{bmatrix} \begin{bmatrix} u_1 \\ i_1 \end{bmatrix}$

ä¸‹é¢ç»™å‡ºçŸ©é˜µä¹˜æ³•$A=A_2 A_1$çš„Pythonæè¿°ã€‚

```python
import numpy as np
from sympy import *

R1 = Symbol('R1', real=True)
R2 = Symbol('R2', real=True)

A1 = np.array([[1,-R1],[0,1]])
A2 = np.array([[1,0],[-1/R2,1]])

A = np.dot(A2,A1)
print('A = \n {}'.format(A))
```

**ã€åŸé¢˜è§£ç­”æ€è·¯ã€‘**

å¯¹äºç”±å•ä¸€ç”µè·¯æ„æˆçš„é›†æˆç”µè·¯ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨`çŸ©é˜µè¿ä¹˜`çš„æ–¹æ³•è¿›è¡Œæ±‚è§£ï¼Œå€¼å¾—æ³¨æ„çš„æ˜¯å¤šä¸ªå•ä¸€ç”µè·¯åˆå¹¶æ—¶çš„`é¡ºåº`ä¸çŸ©é˜µè¿ä¹˜çš„`é¡ºåº`å¯†åˆ‡ç›¸å…³ã€‚äº‹å®ä¸Šï¼Œå¯¹äºå¤æ‚ç”µè·¯ä¹Ÿå¯ä»¥ç”¨ç±»ä¼¼çš„æ–¹æ³•æ±‚è§£ã€‚

- æ±‚æœ€ç»ˆçš„å˜æ¢çŸ©é˜µ

$A = A1 A2 A3 A4 =  A_1 A_2 A_2 A_1$

$\begin{bmatrix} u_{out} \\ i_{out} \end{bmatrix}=A \begin{bmatrix} u_{in} \\ i_{in} \end{bmatrix}$

```python
import numpy as np
from sympy import *

R1 = Symbol('R1', real=True)
R2 = Symbol('R2', real=True)

A1 = np.array([[1,-R1],[0,1]])
A2 = np.array([[1,0],[-1/R2,1]])

A = [A2,A1,A1,A2]
A_OUT = 1
for i in A:
    A_OUT = np.dot(A_OUT,i)

print('A_OUT = \n {}'.format(A_OUT))
```

```python
u0 = Symbol('u0', real=True)
i0 = Symbol('i0', real=True)

IN = np.array([[u0], [i0]]) 

OUT = np.dot(A_OUT, IN)

print('è¾“å‡ºç”µå‹ = {} ä¼ç‰¹'.format(OUT[0][0]))
print('è¾“å‡ºç”µæµ = {} å®‰åŸ¹'.format(OUT[1][0]))
```
